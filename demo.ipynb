```python
figure = plt.figure(figsize=(15, 15))
g = hg.Grid(name="Heatmap", aspect="equal")
g.top("top annotation")
g.top("top dendrogram")
g.top("top title")
g.bottom("bottom annotation")
g.bottom("bottom dendrogram")
g.bottom("bottom title")
g.left("left annotation")
g.left("left dendrogram")
g.left("left title")
g.right("right annotation")
g.right("right dendrogram")
g.right("right title")
g.split("Heatmap", w_ratios=[0.5, 0.5], h_ratios=[0.5, 0.5])
g.plot(figure=figure)
```

# Heatmap Annotation

```python
data = np.random.randint(0, 10, (10, 15))
color_data = [0.1 for _ in range(3)] + [0.2 for _ in range(5)] + [0.3 for _ in range(7)]
```

```python
_, ax = plt.subplots(figsize=(3, .5))

c = hg.Colors(np.array([color_data]), label="Title", label_loc="left")
c.render(ax)
```

```python
h = hg.Heatmap(data)
h.split_col(cut=[5])
h.split_row(cut=[5])
h.add_plot(hg.Colors(np.array([color_data]), label="Title", label_loc="left"), "top", size=.25)
h.add_dendrogram("top")
h.add_dendrogram("left")
h.render()
```
