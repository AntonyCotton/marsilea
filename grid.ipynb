```python
# top
# dir: from bottom to top
# the direction follow the order of how use add more axes
s1 = [3, 4]
s2 = [5, 6]

gridlines = np.concatenate((np.cumsum(s1), np.cumsum(s2)))
gridlines = np.sort(np.unique(gridlines))
print(gridlines)

new_s = ([gridlines[0]] + [i2-i1 for i1, i2 in pairwise(gridlines)])
print(new_s)

it = enumerate(gridlines)
pre_add = 0
ixs = []
for s in s1:
    while True:
        ix, g = next(it)
        if (s + pre_add) == g:
            ixs.append(ix)
            pre_add += s
            break
print("ixs", ixs)

ordered_ixs = len(gridlines) - np.array(ixs) - 1
row_spans = [ixs[0] + 1] + [i2 - i1 for ix, (i1, i2) in
                                    enumerate(pairwise(ixs))]
print("ordered ixs", ordered_ixs)
print("row span", row_spans)
```

```python
# bottom
# dir: from top to bottom
start_ix = 6
s1 = [3, 4]
s2 = [5, 6]

gridlines = np.concatenate((np.cumsum(s1), np.cumsum(s2)))
gridlines = np.sort(np.unique(gridlines))
print(gridlines)

new_s = ([gridlines[0]] + [i2-i1 for i1, i2 in pairwise(gridlines)])
print(new_s)

it = enumerate(gridlines)
pre_add = 0
ixs = []
for s in s1:
    while True:
        ix, g = next(it)
        if (s + pre_add) == g:
            ixs.append(ix)
            pre_add += s
            break
print("ixs", ixs)

row_spans = [ixs[0] + 1] + [i2 - i1 for ix, (i1, i2) in
                                    enumerate(pairwise(ixs))]

ordered_ixs = start_ix + 1 + np.array(ixs)[1::] - (np.asarray(row_spans)[1::] - 1)
ordered_ixs = [start_ix + 1, *ordered_ixs]

print("ordered ixs", ordered_ixs)
print("row span", row_spans)
```
