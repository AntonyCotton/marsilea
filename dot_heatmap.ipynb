```python
%config InlineBackend.figure_format = 'retina'

from hmr import Reloader
import heatgraphy as hg

hg = Reloader(hg)

import numpy as np
import matplotlib.pyplot as plt
```

```python
size = np.random.randint(1, 100, (10, 11))
color = np.random.randint(-10, 10, (10, 11))
matrix = np.random.randint(-10, 10, (10, 11))
```

```python
dh = hg.DotHeatmap(size, color)
dh.add_dendrogram("right")
dh.add_dendrogram("top")
dh.split_col(cut=[5])
dh.split_row(cut=[5])
dh.add_matrix(matrix, cmap="Greens")
dh.render()
```

```python
dh = hg.DotHeatmap(size, color)
dh.add_dendrogram("right")
# dh.add_dendrogram("top")
# dh.split_col(cut=[5])
dh.add_matrix(matrix, cmap="magma")
dh.render()
```
