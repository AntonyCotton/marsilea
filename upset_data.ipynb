```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from matplotlib.collections import PatchCollection, LineCollection
from matplotlib.lines import Line2D

def matrix_plot(matrix, ax=None):
    if ax is None:
        ax = plt.gca()
    
    circles = []
    ls = []
    
    for ix, row in enumerate(matrix):
        lines = []
        for iy, ele in enumerate(row):
            alpha = 0.2
            if ele:
                alpha = 1
                lines.append(iy)
            c = Circle((ix, iy), 0.3, fc=".1", alpha=alpha)
            circles.append(c)
        # draw line
        ymin, ymax = np.min(lines), np.max(lines)
        l = Line2D((ix, ix), (ymin, ymax), color=".1", linewidth=3)
        ls.append(l)
        
    ax.set_xlim(0-0.5, ix+0.5)
    ax.set_ylim(0-0.5, iy+0.5)
    circles = PatchCollection(circles, match_original=True)
    ax.add_artist(circles)
    for l in ls:
        ax.add_artist(l)
    ax.set_aspect(1)
    
                   
            
        
        
    
```

```python
matrix_plot(a)
```
