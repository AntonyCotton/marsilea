```python
import numpy as np
import matplotlib.pyplot as plt
import heatgraphy as hg

data = np.random.randn(10, 10)
labels = (np.arange(11, 21, 1) * 10000000).astype(str)

fig = plt.figure(dpi=150)
h = hg.Heatmap(data)
h.split_row(cut=[5], spacing=0.02)
h.add_plot("right", hg.Labels(labels))
h.add_dendrogram("right")
h.render(figure=fig)
```

```python
h = hg.Heatmap(data)
h.split_row(cut=[5], spacing=0.02)

h.add_plot("right", hg.Labels(labels), size=1, pad=.1)
h.add_dendrogram("right")

h.render()
```

```python
h = hg.Heatmap(data)
#h.split_row(cut=[20], spacing=0.02)
#h.add_dendrogram("right")
# h.add_bar("right", bar_data)
h.add_labels("top", np.arange(10)*100000, size=0.2, pos="center", fontsize=8)
h.add_labels("bottom", np.arange(10)*100000, pos="center", fontsize=8)
h.add_labels("left", np.arange(50), pos="center", fontsize=8)
h.add_labels("right", np.arange(50), pos="center", fontsize=8)
h.render()
```
