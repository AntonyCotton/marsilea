```python
%config InlineBackend.figure_format = 'retina'

from hmr import Reloader
import heatgraphy as hg

hg = Reloader(hg)

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = np.random.randn(10, 10)
bar_data = np.random.randn(10, 10)
```

```python
h = hg.Heatmap(data)
h.split_row(cut=[5])
h.add_plot("right", hg.Bar(bar_data), pad=.1)
h.add_plot("top", hg.Bar(bar_data), pad=.1)
h.add_plot("bottom", hg.Bar(bar_data), pad=.1)
h.add_plot("left", hg.Bar(bar_data), pad=.1)
h.render()
```

```python
h = hg.Heatmap(data, square=True)
h.split_row(cut=[5])
h.add_plot("right", hg.Labels(np.arange(1, 11)), pad=.1)
h.add_plot("top", hg.Bar(bar_data))
h.render()
```

```python
ax = hg.plotter.simple_bar([1, 2, 3, 4], orient="v", fc=".3")
ax.set_axis_off()
```
